```python
from ultralytics import YOLO
```

```python
model = YOLO("../checkpoint/yolov8n_4.pt")
model.export(
    format="onnx",          # 导出格式
    imgsz=[640,1184],              # 输入图像尺寸，可设为 [640, 640] 或 640
    batch=1,                # 批次大小
    dynamic=False,          # 是否支持动态批次（True/False）
    simplify=True,          # 是否简化模型（推荐开启）
    opset=12,               # ONNX 算子版本
    device="cpu"            # 导出设备，"cpu" 或 "cuda:0"
)
```
