```python
import torch
from core.Res18Unet import ResNet18Segmentation as R18U
```

```python
model=R18U(num_classes=2,pretrained=False)
state_dict = torch.load("../checkpoint/Res18Unet_fine_tuning.pth", map_location='cpu')
model.load_state_dict(state_dict)
model.eval();
```

```python
batch_size = 1
channels = 3
height = 300
width = 400
dummy_input = torch.randn(batch_size, channels, height, width)

# 导出 ONNX
onnx_file_path = "Res18Unet_segmentation.onnx"

torch.onnx.export(
    model,                      # 模型
    dummy_input,                # 示例输入
    onnx_file_path,             # 输出路径
    export_params=True,         # 存储训练参数
    opset_version=18,           # ONNX opset 版本
    do_constant_folding=True,   # 常量折叠优化
    input_names=['input'],      # 输入名称
    output_names=['output'],    # 输出名称
    dynamic_axes={
        'input': {0: 'batch_size', 2: 'height', 3: 'width'},  
        'output': {0: 'batch_size', 2: 'height', 3: 'width'}
    }
)

print(f"模型已导出到 {onnx_file_path}")
```
